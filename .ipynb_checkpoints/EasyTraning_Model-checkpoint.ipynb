```python
config = es(
    batch_size=32,
    img_height=224,
    img_width=224,
    data_dir='datasets/chest_xray/',
    validation_split=0.3
)

print('Class names:', config.class_names)
print('Number of classes:', config.num_classes)

base_model = EfficientNetB3

custom_layers = [
    layers.Conv2D(64, (3, 3), activation='relu', padding='same'),
    layers.MaxPooling2D(),
    layers.GlobalAveragePooling2D(),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.5)
]


model = config.create_model(custom_layers,base_model)
model.summary()

epochs = 100

test_loss, test_accuracy, history = config.train_and_evaluate(
    model,
    optimizer='adamw',
    model_save_path='models/' + base_model.__name__ + 'chest_xray' + str(epochs) + '.keras',
    epochs=epochs
)

print(f"Test loss: {test_loss:.4f}")
print(f"Test accuracy: {test_accuracy:.4f}")
```
